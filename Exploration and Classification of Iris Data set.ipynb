### Finding the Probability Density Function(pdf) and Cummumlative Density Function(cdf)

#### PDF:
- Percentage of Distrubtion present for a value
- Probability of that value to be present the set
- Probability graph shows the value and its probability to be present in that area

#### CDF:
- It the sum of previous pdfs to new next pdf

###### For Iris Setosa

```python
counts,bin_edges=np.histogram(setosa["petal.length"],bins=10,density=True)
pdf=counts/(sum(counts))
#print(pdf)
#print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

###### For Iris Virginica

```python
counts,bin_edges=np.histogram(virginica["petal.length"],bins=10,density=True)
pdf=counts/(sum(counts))
#print(pdf)
#print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

# Classification on Dataset

### K Nearest Neighbour Algorithm
- We have used this algorithm as it the most simplest supervised classification and regression algorithm
- We already have the lablels for our data and using this alforithm we can easily classify the variety/species of the algorithm
- We use the KNeighnour Classifier from Scikit-learn Module

### Always Remember:
- Our target is to classify the value based on its features in different species of iris family
- So the values 0, 1 and 2 are the targets and the whole variety column becomes are target class(y)
- On the other hand we require data that can be used to predict this target class and we know that
    - feature(sepal width) + feature(sepal length) + feature(petal width) + feature(petal length) = target(0/1/2)
- So the remaining data columns becomes our data(x)

##### We use the concept y = f(x)
##### or y = x1 + x2 + x3 ......... xN
##### where these x are the features and y is the target of classification

### How to procede and train the KNN Algorithm
![title](knn_howto.png)
1. We split the how data in Training and Test data
(It is important to have a test data to check if are model can be implemented on new data with good acuracy or not)
2. We than fit (train) the model on the training data and training target
3. The scikitlearn module we help us in minimizing the error function automatically and then provide with accurate features to test the model
4. We predict the values on are test data.
5. We then check the accuracy of the model with our test target values and find out how well our model has been trained.

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.model_selection import train_test_split
```

#### Splitting the data into X and Y

```python
y = iris["variety"]
```

```python
x = iris.drop(["variety"], axis=1)
```

```python
y.head()
```

```python
x.head()
```

```python
x.shape
```

```python
y.shape
```

#### Spliiting the data into training and testing 
###### Here we have used 80% data as Training Data and 20% data as Testing Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

#### Training the Data

```python
model = KNN()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
y_test = np.array(y_test)
y_test
```

```python
y_predict
```

# Accuracy of the Model
![title](accuracy.png)

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,y_predict)
print("Accuracy: {:0.2f}%".format(acc))
```
